## Linear Regression Assignment 1

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
dataX = pd.read_csv('Training Data/Linear_X_Train.csv')
dataY = pd.read_csv('Training Data/Linear_Y_Train.csv')
testDataX = pd.read_csv('Testing Data/Linear_X_Test.csv')
testDataX = testDataX.values
dataX = dataX.values
dataY = dataY.values
```

```python
x = dataX
y = dataY
```

```python
x = x.reshape((-1,))
y = y.reshape((-1,))
y = (y-y.mean()) / y.std()
plt.scatter(x, y)
plt.show()
```

```python
x = x.reshape((-1, 1))
y = y.reshape((-1, 1))
```

```python
model = LinearRegression()
# model Fitting
model.fit(x, y)
```

```python
# Model Prediction
output = model.predict(testDataX)
```

```python
y
```

```python
coeff = model.coef_
bias = model.intercept_
print("Coefficient : ",coeff)
print("Intercept : ", bias)
```

```python
plt.scatter(x, y, label="Train data")
plt.plot(testDataX, output, color="red", label="Predicted Data")
plt.legend()
plt.show()
```

```python
# model score
model.score(x, y)
```
