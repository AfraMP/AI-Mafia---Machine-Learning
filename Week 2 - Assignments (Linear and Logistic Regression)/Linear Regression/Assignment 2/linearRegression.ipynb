## Linear Regression Assignment 2

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
trainData = pd.read_csv('./Training Data/Train.csv')
# testDataX = pd.read_csv('./Testing Data/Linear_X_Data.csv')
```

```python
trainData[:5]
```

```python
testDataX = pd.read_csv('./Testing Data/Test.csv')
testDataX = testDataX.values
```

```python
dataY = trainData['target']
dataX = trainData
dataX.drop('target', axis = 1, inplace=True)
trainDataX = dataX.values
trainDataY = dataY.values
trainDataY = (trainDataY - trainDataY.mean()) / trainDataY.std()
trainDataY.shape, trainDataX.shape
```

```python
plt.scatter(trainDataX[:, 0], trainDataY)
plt.scatter(trainDataX[:, 1], trainDataY, color='g')
plt.scatter(trainDataX[:, 2], trainDataY, color='r')
plt.scatter(trainDataX[:, 3], trainDataY, color='y')
plt.ylabel('features')
plt.xlabel('target')
```

```python
model = LinearRegression()
model.fit(trainDataX, trainDataY)
```

```python
output = model.predict(testDataX)
```

```python
coeff = model.coef_
bias = model.intercept_
print("Coefficient : ", coeff)
print("Intercept : ", bias)
```
