## Logistic Regression Assignment 1

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import log_loss
```

```python
trainDataX = pd.read_csv('./Training Data/Logistic_X_Train.csv')
trainDataY = pd.read_csv('./Training Data/Logistic_Y_Train.csv')
testDataX = pd.read_csv('./Testing Data/Logistic_X_Test.csv')
trainDataX
```

```python
trainDataX = trainDataX.values
# trainDataX = trainDataX.reshape((-1, ))
trainDataY = trainDataY.values
trainDataY = trainDataY.reshape((-1, ))
trainDataX
```

```python
model = LogisticRegression()
model.fit(trainDataX, trainDataY)
```

```python
testDataX = testDataX.values
output = model.predict(testDataX)
```

```python
model.score(trainDataX, trainDataY)
# trainDataX
```
